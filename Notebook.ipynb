<h4>C. Detection des valeurs aberrantes :</h4>

<h5>Approche mtier : </h5>

En effectuant de recherches nous pouvons facilement trouver la plus grande taille d'un arbre atteinte a ce jour ainsi la circonfrence max du tronc nous pouvons donc utiliser ces valeurs afin de filtrer les valeurs aberrantes pour la hauteur et la circonfrence.

<ul>Sources : 
    <li><a href='https://www.futura-sciences.com/planete/questions-reponses/arbre-hauteur-maximale-arbre-peut-il-atteindre-13058/'>Hauteur max : 120 mtres</a></li>
    <li><a href='http://www.sequoias.eu/Pages/hauteur_maxi.htm#:~:text=Le%20seuil%20de%20hauteur%20maximale,138m%20selon%20les%20deux%20%C3%A9tudes.&text=1%20Le%20%22Cary%20Fir%22%2C,la%20mesure%20est%20cependant%20contest%C3%A9e.'>Circonfrence max: 30 mtres</a></li>

```python
# Dtermination des bornes suprieures et infrieures pour la hauteur
born_inf_hauteur = 0  # Valeur minimale acceptable pour la hauteur
born_sup_hauteur = 120  # Valeur maximale acceptable pour la hauteur

# Dtermination des bornes suprieures et infrieures pour la circonfrence
born_inf_circonf = 0  # Valeur minimale acceptable pour la circonfrence
born_sup_circonf = 3000  # Valeur maximale acceptable pour la circonfrence

# Filtrage des valeurs aberrantes dans la colonne 'hauteur_m'
outliers_hauteur = data[(data['hauteur_m'] < born_inf_hauteur) | (data['hauteur_m'] > born_sup_hauteur)]

# Filtrage des valeurs aberrantes dans la colonne 'circonference_cm'
outliers_circonf = data[(data['circonference_cm'] < born_inf_circonf) | (data['circonference_cm'] > born_sup_circonf)]

print("Nombre de valeurs aberrantes dans la colonne 'hauteur_m' :", len(outliers_hauteur))
print("Nombre de valeurs aberrantes dans la colonne 'circonference_cm' :", len(outliers_circonf))

```

Nous crons un DataFrame "outliers values" qui regroupe nos deux colonnes afin de vrifier si notre filtre a bien fonctionn : 

```python
outliers_values = pd.DataFrame({'hauteur_m': outliers_hauteur['hauteur_m'], 'circonference_cm': outliers_circonf['circonference_cm']})
```

```python
outliers_values
```

Nous pouvons observer des valeurs NaN dans notre DataFrame "outliers values" alors qu'auparavant nous n'avions pas de NaN dans notre DataFrame "data", cela s'explique car dans le DataFrame "outliers values" les valeurs dans les colonnes "hauteur m" et "circonfrence" qui dpassent les bornes suprieures ou infrieures des filtres peuvent tre remplacs par des NaN lors de l'extraction.

Nous pouvons donc dduire que nos filtres ont bien fonctionn.

<h5>Approche statistique (mthode interquartile) : </h5>

```python
# Calculer les quartiles Q1 et Q3 pour la hauteur et la circonfrence
Q1_height = data['hauteur_m'].quantile(0.25)
Q3_height = data['hauteur_m'].quantile(0.75)

Q1_circumference = data['circonference_cm'].quantile(0.25)
Q3_circumference = data['circonference_cm'].quantile(0.75)

# Calculer la plage interquartile (IQR)
IQR_height = Q3_height - Q1_height
IQR_circumference = Q3_circumference - Q1_circumference

# Dfinir les bornes suprieures et infrieures
upper_bound_height = Q3_height + 1.5 * IQR_height
lower_bound_height = Q1_height - 1.5 * IQR_height

upper_bound_circumference = Q3_circumference + 1.5 * IQR_circumference
lower_bound_circumference = Q1_circumference - 1.5 * IQR_circumference

# Filtrer les valeurs aberrantes
outliers_height = data[(data['hauteur_m'] < lower_bound_height) | (data['hauteur_m'] > upper_bound_height)]
outliers_circumference = data[(data['circonference_cm'] < lower_bound_circumference) | (data['circonference_cm'] > upper_bound_circumference)]

print("Nombre de valeurs aberrantes dans la colonne 'hauteur_m' (approche interquartile) :", len(outliers_height))
print("Nombre de valeurs aberrantes dans la colonne 'circonference_cm' (approche interquartile) :", len(outliers_circumference))
```

<h5>Approche statistique (mthode Z score) : </h5>

<h4>D. Gestion des valeurs aberrantes :</h4>

```python
# Filtrage des valeurs aberrantes dans la colonne 'hauteur_m'
data = data[(data['hauteur_m'] >= born_inf_hauteur) & (data['hauteur_m'] <= born_sup_hauteur)]

# Filtrage des valeurs aberrantes dans la colonne 'circonference_cm'
data = data[(data['circonference_cm'] >= born_inf_circonf) & (data['circonference_cm'] <= born_sup_circonf)]

# Affichage du DataFrame aprs nettoyage des valeurs aberrantes
data
```

<h5>Graphique 1 :</h5>

Affiche le nombre d'arbres par espce dans chaque arrondissement.

```python
# Slection des colonnes pertinentes
arbres_selectionnes = data[['arrondissement', 'libelle_francais', 'espece']]

# Filtrage des donnes pour exclure les lignes avec des valeurs manquantes
arbres_filtres = arbres_selectionnes.dropna(subset=['arrondissement', 'libelle_francais', 'espece'])

# Opration d'agrgation pour compter le nombre d'arbres par combinaison d'arrondissement, libell franais et espce
arbres_agreges = arbres_filtres.groupby(['arrondissement', 'libelle_francais', 'espece']).size().reset_index(name='count')

# Opration d'agrgation supplmentaire pour calculer le nombre total d'arbres par arrondissement
arbres_par_arrondissement = arbres_agreges.groupby('arrondissement')['count'].sum().reset_index(name='total')

# Jointure des deux DataFrames pour obtenir le pourcentage des espces dans chaque arrondissement
pourcentages_especes = arbres_agreges.merge(arbres_par_arrondissement, on='arrondissement')
pourcentages_especes['percentage'] = (pourcentages_especes['count'] / pourcentages_especes['total']) * 100
```

```python
# Cration du diagramme avec Seaborn
plt.figure(figsize=(8, 8))
sns.countplot(data=pourcentages_especes, x='arrondissement', hue='espece')
plt.title('Nombre d\'arbres par espce dans chaque arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Nombre d\'arbres')
plt.legend(title='Espce', loc='best')
plt.xticks(rotation=45)
plt.show()
```

Nous pouvons observer que le nombre d'espces et d'arrondissements est trop lev pour afficher un diagramme qui montre le nombre d'arbres par espce dans chaque arrondissement, une alternative est de de se concentrer sur les espces les plus courantes ou les plus reprsentatives.

```python
# Slection des deux espces les plus courantes
top_2_especes = pourcentages_especes.groupby('espece')['count'].sum().nlargest(2).index

# Filtrage du DataFrame pour inclure seulement les deux espces les plus courantes
data_filtre = pourcentages_especes[pourcentages_especes['espece'].isin(top_2_especes)]
```

```python
# Cration du graphique en barres sans les barres d'intervalle de confiance
plt.figure(figsize=(10, 8))
sns.set_palette('pastel')
ax = sns.barplot(data=data_filtre, x='arrondissement', y='count', hue='espece', errorbar=None)
plt.title('Nombre d\'arbres par espce dans chaque arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Nombre d\'arbres')
plt.legend(title='Espces', loc='best')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  # Affichage des noms des arrondissements  la verticale
plt.show()
```
