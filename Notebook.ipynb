<h4>C. Detection des valeurs aberrantes :</h4>

<h5>Approche métier : </h5>

En effectuant de recherches nous pouvons facilement trouver la plus grande taille d'un arbre atteinte a ce jour ainsi la circonférence max du tronc nous pouvons donc utiliser ces valeurs afin de filtrer les valeurs aberrantes pour la hauteur et la circonférence.

<ul>Sources : 
    <li><a href='https://www.futura-sciences.com/planete/questions-reponses/arbre-hauteur-maximale-arbre-peut-il-atteindre-13058/'>Hauteur max : 120 mètres</a></li>
    <li><a href='http://www.sequoias.eu/Pages/hauteur_maxi.htm#:~:text=Le%20seuil%20de%20hauteur%20maximale,138m%20selon%20les%20deux%20%C3%A9tudes.&text=1%20Le%20%22Cary%20Fir%22%2C,la%20mesure%20est%20cependant%20contest%C3%A9e.'>Circonférence max: 30 mètres</a></li>

```python
# Détermination des bornes supérieures et inférieures pour la hauteur
born_inf_hauteur = 0  # Valeur minimale acceptable pour la hauteur
born_sup_hauteur = 120  # Valeur maximale acceptable pour la hauteur

# Détermination des bornes supérieures et inférieures pour la circonférence
born_inf_circonf = 0  # Valeur minimale acceptable pour la circonférence
born_sup_circonf = 3000  # Valeur maximale acceptable pour la circonférence

# Filtrage des valeurs aberrantes dans la colonne 'hauteur_m'
outliers_hauteur = data[(data['hauteur_m'] < born_inf_hauteur) | (data['hauteur_m'] > born_sup_hauteur)]

# Filtrage des valeurs aberrantes dans la colonne 'circonference_cm'
outliers_circonf = data[(data['circonference_cm'] < born_inf_circonf) | (data['circonference_cm'] > born_sup_circonf)]

print("Nombre de valeurs aberrantes dans la colonne 'hauteur_m' :", len(outliers_hauteur))
print("Nombre de valeurs aberrantes dans la colonne 'circonference_cm' :", len(outliers_circonf))

```

Nous créons un DataFrame "outliers values" qui regroupe nos deux colonnes afin de vérifier si notre filtre a bien fonctionné : 

```python
outliers_values = pd.DataFrame({'hauteur_m': outliers_hauteur['hauteur_m'], 'circonference_cm': outliers_circonf['circonference_cm']})
```

```python
outliers_values
```

Nous pouvons observer des valeurs NaN dans notre DataFrame "outliers values" alors qu'auparavant nous n'avions pas de NaN dans notre DataFrame "data", cela s'explique car dans le DataFrame "outliers values" les valeurs dans les colonnes "hauteur m" et "circonférence" qui dépassent les bornes supérieures ou inférieures des filtres peuvent être remplacés par des NaN lors de l'extraction.

Nous pouvons donc déduire que nos filtres ont bien fonctionné.

<h5>Approche statistique (méthode interquartile) : </h5>

```python
# Calculer les quartiles Q1 et Q3 pour la hauteur et la circonférence
Q1_height = data['hauteur_m'].quantile(0.25)
Q3_height = data['hauteur_m'].quantile(0.75)

Q1_circumference = data['circonference_cm'].quantile(0.25)
Q3_circumference = data['circonference_cm'].quantile(0.75)

# Calculer la plage interquartile (IQR)
IQR_height = Q3_height - Q1_height
IQR_circumference = Q3_circumference - Q1_circumference

# Définir les bornes supérieures et inférieures
upper_bound_height = Q3_height + 1.5 * IQR_height
lower_bound_height = Q1_height - 1.5 * IQR_height

upper_bound_circumference = Q3_circumference + 1.5 * IQR_circumference
lower_bound_circumference = Q1_circumference - 1.5 * IQR_circumference

# Filtrer les valeurs aberrantes
outliers_height = data[(data['hauteur_m'] < lower_bound_height) | (data['hauteur_m'] > upper_bound_height)]
outliers_circumference = data[(data['circonference_cm'] < lower_bound_circumference) | (data['circonference_cm'] > upper_bound_circumference)]

print("Nombre de valeurs aberrantes dans la colonne 'hauteur_m' (approche interquartile) :", len(outliers_height))
print("Nombre de valeurs aberrantes dans la colonne 'circonference_cm' (approche interquartile) :", len(outliers_circumference))
```

<h5>Approche statistique (méthode Z score) : </h5>

<h4>D. Gestion des valeurs aberrantes :</h4>

Maintenant que nous avons identifié les valeurs aberrantes dans les colonnes "hauteur m" et "circonférence" à l'aide des bornes supérieures et inférieures spécifiées, nous pouvons les nettoyer en les supprimant du DataFrame data.

```python
# Filtrage des valeurs aberrantes dans la colonne 'hauteur_m'
data = data[(data['hauteur_m'] >= born_inf_hauteur) & (data['hauteur_m'] <= born_sup_hauteur)]

# Filtrage des valeurs aberrantes dans la colonne 'circonference_cm'
data = data[(data['circonference_cm'] >= born_inf_circonf) & (data['circonference_cm'] <= born_sup_circonf)]

# Affichage du DataFrame après nettoyage des valeurs aberrantes
data
```

On vient vérifier si tout c'est bien dérouler et si nous n'avons plus de Nan ou de Valeurs manquante après notre filtrage et si celui-ci a bien fonctionné :

```python
print("Nombre de valeurs manquantes dans la colonne 'circonference_cm':", data['circonference_cm'].isnull().sum())
print("Nombre de valeurs manquantes dans la colonne 'hauteur_m':", data['hauteur_m'].isnull().sum())
print("Valeurs minimales et maximales pour la hauteur_m :", data['hauteur_m'].min(), "-", data['hauteur_m'].max())
print("Valeurs minimales et maximales pour la circonference_cm :", data['circonference_cm'].min(), "-", data['circonference_cm'].max())
```

<h4>Analyse bivariée : </h4>

<h5>Graphique 1 :</h5>

Affiche le nombre d'arbres par espèce dans chaque arrondissement.

```python
# Sélection des colonnes pertinentes
arbres_selectionnes = data[['arrondissement', 'libelle_francais', 'espece']]

# Filtrage des données pour exclure les lignes avec des valeurs manquantes
arbres_filtres = arbres_selectionnes.dropna(subset=['arrondissement', 'libelle_francais', 'espece'])

# Opération d'agrégation pour compter le nombre d'arbres par combinaison d'arrondissement, libellé français et espèce
arbres_agreges = arbres_filtres.groupby(['arrondissement', 'libelle_francais', 'espece']).size().reset_index(name='count')

# Opération d'agrégation supplémentaire pour calculer le nombre total d'arbres par arrondissement
arbres_par_arrondissement = arbres_agreges.groupby('arrondissement')['count'].sum().reset_index(name='total')

# Jointure des deux DataFrames pour obtenir le pourcentage des espèces dans chaque arrondissement
pourcentages_especes = arbres_agreges.merge(arbres_par_arrondissement, on='arrondissement')
pourcentages_especes['percentage'] = (pourcentages_especes['count'] / pourcentages_especes['total']) * 100
```

```python
# Création du diagramme avec Seaborn
plt.figure(figsize=(8, 8))
sns.countplot(data=pourcentages_especes, x='arrondissement', hue='espece')
plt.title('Nombre d\'arbres par espèce dans chaque arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Nombre d\'arbres')
plt.legend(title='Espèce', loc='best')
plt.xticks(rotation=45)
plt.show()
```

Nous pouvons observer que le nombre d'espèces et d'arrondissements est trop élevé pour afficher un diagramme qui montre le nombre d'arbres par espèce dans chaque arrondissement, une alternative est de de se concentrer sur les espèces les plus courantes ou les plus représentatives.

```python
# Sélection des deux espèces les plus courantes
top_2_especes = pourcentages_especes.groupby('espece')['count'].sum().nlargest(2).index

# Filtrage du DataFrame pour inclure seulement les deux espèces les plus courantes
data_filtre = pourcentages_especes[pourcentages_especes['espece'].isin(top_2_especes)]
```

```python
# Création du graphique en barres sans les barres d'intervalle de confiance
plt.figure(figsize=(10, 8))
sns.set_palette('pastel')
ax = sns.barplot(data=data_filtre, x='arrondissement', y='count', hue='espece', errorbar=None)
plt.title('Nombre d\'arbres par espèce dans chaque arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Nombre d\'arbres')
plt.legend(title='Espèces', loc='best')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  # Affichage des noms des arrondissements à la verticale
plt.show()
```

En étudiant plus en profondeur le problème du voyageur de commerce , j'ai pus observer que l'une des méthodes de résolution de celui- ci était l'algorithme de colonies de fourmis.

Au cours de l'approfondissement de mes investigations, j'ai découvert une publication scientifique qui éclaire pertinemment la problématique à l'étude.

Cette étude utilise des réseaux de neurones en rétro-propagation afin de pouvoir résoudre le problème du voyageur de commerce.
Toutefois, l'application des réseaux de neurones pour résoudre ce problème a également été étudiée.

Par exemple, dans le cas du problème de routage de véhicules (VRP), l'approche hybride qui utilise le réseau de neurones auto-organisateur (SOM) comme une recherche locale dans un algorithme mémétique basé sur la population a été présentée


<ul>Sources : 
    <li><a href='https://www.semanticscholar.org/paper/From-the-TSP-to-the-Dynamic-VRP%3A-An-Application-of-Hajjam-Cr%C3%A9put/57abb85f3833eeeeb190b48945d0f68a0c7a1bc6'>Du TSP au VRP dynamique : une application des réseaux de neurones dans la métaheuristique basée sur la population</a></li>
</ul>

Mais aussi avec la methode de l'algorithme génétique.

Cette méthode, inspirée du processus de sélection naturelle, recherche un espace pour une solution approximative à des problèmes avec de multiples solutions.

Un exemple d'application est la recherche de l'itinéraire optimal dans le cadre du problème VRP

<ul>Sources : 
    <li><a href='https://www.mdpi.com/2079-9292/10/24/3147'>Algorithmes génétiques sélectionnés pour la résolution de problèmes de routage de véhicules</a></li>
</ul>


Cependant, il convient de souligner que la résolution de ce problème peut se faire de plusieurs façons et ne nécessite pas nécessairement l'implémentation d'un réseau de neurones.

Parallèlement, des méthodes mathématiques ont été largement utilisées pour résoudre le problème du voyageur de commerce (TSP) et le VRP.

Le TSP, par exemple, a été formulé pour la première fois comme un problème de trouver l'itinéraire le plus court à travers un graphe qui visite chaque nœud  exactement une fois.

Plusieurs techniques d'optimisation importantes ont été développées à partir de l'étude du TSP, notamment les méthodes de plans de coupe et les algorithmes de branch and bound.

<ul>Sources : 
    <li><a href='https://ar5iv.labs.arxiv.org/html/2105.10950'>Un examen des approches de modélisation des problèmes de routage de véhicules appliqués</a></li>
</ul>

Dans le cas du VRP, la notion de capacité est directement impliquée, où la demande en ressources fixes des clients d'une part, et le nombre de clients qui peuvent être desservis par un seul véhicule d'autre part, sont contraints par la capacité du véhicule.

Cela a introduit à la fois la composante de routage (similaire au TSP) et la composante d'attribution (les clients doivent être partitionnés en plusieurs routes)

<ul>Sources : 
    <li><a href='https://ar5iv.labs.arxiv.org/html/2105.10950'>A review of approaches to modeling applied vehicle routing problems</a></li>
</ul>
